```python
car_sales_missing_data=pd.read_csv('/content/car-sales-missing-data.csv')
car_sales_missing_data
```

```python
car_sales_missing_data.isna().sum()
```

```python
car_sales_missing_data.isnull().sum()
```

```python
car_sales_missing_data.dropna(subset=['Price'],inplace=True)
```

```python
car_sales_missing_data.isnull().sum()
```

```python
car_sales_missing_data['Odometer']=car_sales_missing_data['Odometer'].fillna(
    car_sales_missing_data['Odometer'].mean())
```

```python
car_sales_missing_data
```

```python
car_sales_missing_data.isnull().sum()
```

```python
car_sales_missing_data['Colour']=car_sales_missing_data['Colour'].fillna(
    car_sales_missing_data['Colour'].mode()[0]
)

car_sales_missing_data['Make']=car_sales_missing_data['Make'].fillna(
    car_sales_missing_data['Make'].mode()[0]
)
```

```python
car_sales_missing_data
```

```python
for col in ['Make','Colour']:
  car_sales_missing_data[col]=car_sales_missing_data[col].fillna(
      car_sales_missing_data[col].mode()[0]
  )
```

```python
table_a=pd.DataFrame({'id':['a1','a2','a3','a4'],'Value':['A','B','C','AA']})
table_b=pd.DataFrame({'id':['a1','a2','a3','a5'],'Value':['D','E','F','BB']})
pd.merge(table_a,table_b,on='id')  #inner join
```

```python
table_a=pd.DataFrame({'id1':['a1','a2','a3','a4'],'Value':['A','B','C','AA']})
table_b=pd.DataFrame({'id2':['a1','a2','a3','a5'],'Value':['D','E','F','BB']})
pd.merge(table_a,table_b,left_on='id1',right_on='id2',how='right')
```
